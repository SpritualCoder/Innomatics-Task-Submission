Problem 1: Social Media - Post Engagement Analyzer

```python
# Problem 1: Social Media - Post Engagement Analyzer

def analyze_engagement(likes_list):
    # Initialize total likes to 0
    total_likes = 0

    # Loop through each like count in the list
    for like in likes_list:
        total_likes += like   # Add each like to total

    # Check if post is viral
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display results
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example data (likes from different sources)
likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker

def rainfall_checker(rainfall_data, required_level):
    # Calculate total rainfall
    total = 0
    for rain in rainfall_data:
        total += rain

    # Calculate average rainfall
    average = total / len(rainfall_data)

    # Compare with required rainfall level
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Display results
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example rainfall data
rainfall = [70, 75, 80, 65]
rainfall_checker(rainfall, 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicates(usernames):
    # Convert list to set (removes duplicates automatically)
    unique_users = set(usernames)

    # If length changes, duplicates exist
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usernames
users = ["raj", "amit", "raj", "neha"]
detect_duplicates(users)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    # Create empty list to store premium crops
    premium = []

    # Loop through price list
    for price in prices:
        # Check if price is greater than 2000
        if price > 2000:
            premium.append(price)  # Add to premium list

    # Display premium crop prices
    print("Premium Crops:", premium)


# Example crop prices
crop_prices = [1500, 2500, 3200, 1800]
filter_premium_crops(crop_prices)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Problem 9: E-Learning – Student Attendance Eligibility System

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)  # Total classes conducted
    attended = 0  # Counter for attended classes

    # Count number of classes attended
    for status in attendance_list:
        if status == 1:   # 1 means Present
            attended += 1

    # Calculate attendance percentage
    percentage = (attended / total_classes) * 100

    # Check eligibility condition
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Display results
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


# Example attendance record
attendance = [1, 1, 1, 0, 1]
attendance_checker(attendance)
```

Problem 10: Smart Electricity Bill Analyzer

```python
# Problem 10: Smart Electricity Bill Analyzer

def electricity_bill(units):
    bill = 0  # Initialize bill amount

    # Apply slab-based billing
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify usage based on bill amount
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    # Display bill details
    print("Total Bill:", bill)
    print("Usage Level:", usage)


# Example
electricity_bill(250)
```
