1: Mobile Recharge Validation System

```python
# Problem 1: Mobile Recharge Validation System

def validateRecharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        return False
    else:
        print("Recharge Successful!")
        return True


# Retry system using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        if validateRecharge(amount):
            break
    except ValueError:
        print("Please enter a valid number.")
```

2: Inventory Reorder Alert System

```python
# Problem 2: Inventory Reorder Alert System

def checkInventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 5,
    "Monitor": 18
}

checkInventory(inventory)
```

3: Student Result Processing System

```python
# Problem 3: Student Result Processing System

def processResult(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 45, 70, 55]
result = processResult(marks_list)

print("Result:", result)
```

4: Cab Fare Estimator

```python
# Problem 4: Cab Fare Estimator

def calculateFare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak:
        fare = fare + (fare * 0.25)

    return fare


# Retry option using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peakInput = input("Is it peak hour? (yes/no): ").lower()

        peak = True if peakInput == "yes" else False

        totalFare = calculateFare(distance, peak)
        print("Total Fare: ₹", totalFare)

        break
    except ValueError:
        print("Invalid input. Please try again.")
```

5: Attendance Eligibility Checker

```python
# Problem 5: Attendance Eligibility Checker

def checkEligibility(attendance):
    present_count = 0

    for day in attendance:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P"]
status = checkEligibility(attendance_list)

print("Attendance Status:", status)
```

6: Password Strength Checker

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and special character @#$)"


user_password = input("Enter password: ")
result = check_password_strength(user_password)

print(result)
```
